<i> The figure above shows the correlation among the independent variable whch is our focus

```python
def corre(df, thre):
    corr_cols = set()
    corr = df.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > thre:
                cols = corr.columns[i]
                corr_cols.add(cols)
                
    return corr_cols
```

```python
corre(X_train, 0.75)
```

<b>Breakdown of the corre function

df = The Dataset<br>thre = The Threshold ypu want to set the correletion to

<i>I created a variable(corr_cols) to accept values as set(Since set stores unique values)<br>corr: To Store the correlation values of the dataset<br>Then i looped through the columns of the dataset to get the correlation of an individual column to the other columns<br> After this Process I checked it with threshold i set my Correleraion value to be<br> If it meets it, it gives am output which i can decided to do away with from the dataset
